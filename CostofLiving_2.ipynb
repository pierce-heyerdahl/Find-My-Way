```python
draftData['URBAN AREA AND STATE'] = draftData['URBAN AREA AND STATE'].str[:-3]
draftData.rename(columns = {'URBAN AREA AND STATE':'CITY'}, inplace=True)
draftData.head()
```

```python
draftData = draftData.assign(CITY = draftData.CITY.str.split("-"))
draftData = draftData.explode('CITY')
draftData.head()
```

```python
finalData = draftData[['CITY','STATE','STATE(abbr)','COMPOSITE INDEX']]
finalData.reset_index(drop = True, inplace = True)
finalData.drop(finalData.index[-1], inplace = True)
finalData.head()
```

```python
# finalData.to_excel('CostofLivingData.xlsx', index = False)
```

```python
stateAve = finalData.groupby(['STATE','STATE(abbr)'])['COMPOSITE INDEX'].mean().round(2).reset_index()
stateAve.head()
```
