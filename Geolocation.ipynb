```python
# Abstract data that meet the criteria.
USdata = data[(data['country'] == 'United States')]
USdata.head()
```

```python
# Select columns we need and rename columns.
finalData = pd.DataFrame(USdata, columns = ['city', 'admin_name', 'lat', 'lng'])
finalData.rename(columns = {'city':'CITY', 'admin_name':'STATE'}, inplace=True)
finalData.head()
```

```python
# re-order columns.
finalData = finalData[['CITY', 'STATE','STATE(abbr)','lat','lng']]
finalData.head()
```
